```python
import cv2
import numpy as np


def image_processing(image):

    #Grayscaling
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # detect circles

    #Median Blurr for better results
    blur = cv2.medianBlur(cv2.cvtColor(gray, cv2.COLOR_RGB2GRAY), 5)

    invert = cv2.bitwise_not(blur)

    _,thresholded_image= cv2.threshold(invert,150,255,cv2.THRESH_BINARY)
    cv2.imshow('thresholded',thresholded_image)
    cv2.waitKey(0)

    return thresholded_image
    


def extract_roi(image):
    #Using hough circles to detect circles
    circles = cv2.HoughCircles(image, cv2.HOUGH_GRADIENT, 1, 20, param1=200, param2=20, minRadius=0, maxRadius=0)

    #Converting all the hough circle output into int data type
    circles = np.uint16(np.around(circles))

    # draw mask
    black_image = np.full((image.shape[0], image.shape[1]), 0, dtype=np.uint8)

    #cv2.imshow('Black Image',black_image)
    #cv2.waitKey(0)

    for i in circles[0, :]:
    
        #Drawing the filled white circle using hough circle output
        cv2.circle(black_image, (i[0], i[1]), i[2]+5, (255, 255, 255), -1)
    
    #cv2.imshow("Mask (Black image with a filled white circle)",black_image)
    #cv2.waitKey(0)


    # Applying the black mask over the real image with a trasparent circular window 
    roi = cv2.bitwise_or(image,image, mask=black_image)
    
    return roi

    
# load the image
img = cv2.imread('eye.jpg')
cv2.imshow('Eye',img)
cv2.waitKey(0)
processed_image = image_processing(img)    
ROI = extract_roi(processed_image)
cv2.imshow('Region of interest',ROI)
cv2.waitKey(0)

cv2.destroyAllWindows()
```
